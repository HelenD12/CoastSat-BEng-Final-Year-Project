```python
import pandas as pd
import os
import re

# Update file path to folder containing data
data_folder = r'C:\Users\helen\Documents\CCO'

# Define file list and desired Reg IDs
file_list = [
    '6bMU25-2_20070517tp.txt',
    '6bMU25-2_20070517tp.txt',
    '6bMU25-2_20071028tp.txt',
    '6bMU25-2_20080223tp.txt',
    '6bMU25-2_20080902tp.txt',
    '6bMU25-2_20090212tip.txt',
    '6bMU25-2_20090921tip.txt',
    '6bMU25-2_20100429tip.txt',
    '6bMU25-2_20101010tip.txt',
    '6bMU25-2_20110219tip.txt',
    '6bMU25-3_20070927tp.txt',
    '6bMU25-3_20080407tip.txt',
    '6bMU25-3_20090211tip.txt',
    '6bMU25-3_20100428tip.txt',
    '6bMU25-3_20100812tp.txt',
    '6bMU25-3_20110219tip.txt',
    '6bMU26-1_20070930tp.txt',
    '6bMU26-1_20080407tip.txt',
    '6bMU26-1_20080421tpsp.txt',
    '6bMU26-1_20080909tpsp.txt',
    '6bMU26-1_20081016tip.txt',
    '6bMU26-1_20090115tpsp.txt',
    '6bMU26-1_20090211tip.txt',
    '6bMU26-1_20090919tip.txt',
    '6bMU26-1_20091022tpsp.txt',
    '6bMU26-1_20100428tip.txt',
    '6bMU26-1_20100813tp.txt',
    '6bMU26-1_20101206tip.txt',
    '6bMU26-1_20110219tip.txt',
    '6bMU26-2_20070930tp.txt',
    '6bMU26-2_20080408tip.txt',
    '6bMU26-2_20080908tpsp.txt',
    '6bMU26-2_20081017tip.txt',
    '6bMU26-2_20090114tpsp.txt',
    '6bMU26-2_20090212tip.txt',
    '6bMU26-2_20090920tip.txt',
    '6bMU26-2_20091022tpsp.txt',
    '6bMU26-2_20100329tip.txt',
    '6bMU26-2_20100814tp.txt',
    '6bMU26-2_20101207tip.txt',
    '6bMU26-2_20110220tip.txt',
    '6bMU26-3_20070928tp.txt',
    '6bMU26-3_20080408tip.txt',
    '6bMU26-3_20090114tpsp.txt',
    '6bMU26-3_20090212tip.txt',
    '6bMU26-3_20100329tip.txt',
    '6bMU26-3_20100812tp.txt',
    '6bMU26-3_20110220tip.txt',
    '6bSU25-2_20111221tip.txt',
    '6bSU25-2_20120807tip.txt',
    '6bSU25-2_20120807tp.txt',
    '6bSU25-2_20121102tip.txt',
    '6bSU25-2_20130529tip.txt',
    '6bSU25-2_20140303tip.txt',
    '6bSU25-2_20141029tip.txt',
    '6bSU25-2_20150205tip.txt',
    '6bSU25-2_20160222tip.txt',
    '6bSU25-2_20161115tip.txt',
    '6bSU25-2_20170714tp.txt',
    '6bSU25-2_20171106tip.txt',
    '6bSU25-2_20180306tip.txt',
    '6bSU25-2_20180614tp.txt',
    '6bSU25-2_20181009tip.txt',
    '6bSU25-2_20190322tip.txt',
    '6bSU25-2_20190928tip.txt',
    '6bSU25-2_20200213tip.txt',
    '6bSU25-2_20200904tp.txt',
    '6bSU25-2_20201116tip.txt',
    '6bSU25-2_20210315tip.txt',
    '6bSU25-2_20210529tp.txt',
    '6bSU25-2_20211105tip.txt',
    '6bSU25-2_20220303tp.txt',
    '6bSU25-2_20221123tip.txt',
    '6bSU25-2_20230221tip.txt',
    '6bSU25-2_20230420tp.txt',
    '6bSU25-2_20231004tip.txt',
    '6bSU26-1_20111129tip.txt',
    '6bSU26-1_20120804tip.txt',
    '6bSU26-1_20120804tp.txt',
    '6bSU26-1_20121020tip.txt',
    '6bSU26-1_20130528tip.txt',
    '6bSU26-1_20130911tp.txt',
    '6bSU26-1_20140319tip.txt',
    '6bSU26-1_20140815tp.txt',
    '6bSU26-1_20141107tip.txt',
    '6bSU26-1_20141211tpsp.txt',
    '6bSU26-1_20150319tip.txt',
    '6bSU26-1_20150618tp.txt',
    '6bSU26-1_20151216tip.txt',
    '6bSU26-1_20160108tpsp.txt',
    '6bSU26-1_20160323tip.txt',
    '6bSU26-1_20160708tp.txt',
    '6bSU26-1_20161115tip.txt',
    '6bSU26-1_20170714tp.txt',
    '6bSU26-1_20171107tip.txt',
    '6bSU26-1_20180306tip.txt',
    '6bSU26-1_20180517tp.txt',
    '6bSU26-1_20181009tip.txt',
    '6bSU26-1_20190322tip.txt',
    '6bSU26-1_20190605tp.txt',
    '6bSU26-1_20190928tip.txt',
    '6bSU26-1_20200128tip.txt',
    '6bSU26-1_20200709tp.txt',
    '6bSU26-1_20201118tip.txt',
    '6bSU26-1_20210315tip.txt',
    '6bSU26-1_20210715tp.txt',
    '6bSU26-1_20211105tip.txt',
    '6bSU26-1_20220803tp.txt',
    '6bSU26-1_20221123tip.txt',
    '6bSU26-1_20230221tip.txt',
    '6bSU26-1_20230520tp.txt',
    '6bSU26-1_20231004tip.txt',
    '6bSU26-2_20111215tip.txt',
    '6bSU26-2_20120806tip.txt',
    '6bSU26-2_20120806tp.txt',
    '6bSU26-2_20121031tip.txt',
    '6bSU26-2_20130523tip.txt',
    '6bSU26-2_20130824tp.txt',
    '6bSU26-2_20131203tip.txt',
    '6bSU26-2_20140109tpsp.txt',
    '6bSU26-2_20140304tip.txt',
    '6bSU26-2_20140812tp.txt',
    '6bSU26-2_20141028tip.txt',
    '6bSU26-2_20150307tip.txt',
    '6bSU26-2_20150507tp.txt',
    '6bSU26-2_20160114tip.txt',
    '6bSU26-2_20160721tp.txt',
    '6bSU26-2_20161116tip.txt',
    '6bSU26-2_20170713tp.txt',
    '6bSU26-2_20171106tip.txt',
    '6bSU26-2_20180306tip.txt',
    '6bSU26-2_20180517tp.txt',
    '6bSU26-2_20181010tip.txt',
    '6bSU26-2_20190322tip.txt',
    '6bSU26-2_20190606tp.txt',
    '6bSU26-2_20190929tip.txt',
    '6bSU26-2_20200129tip.txt',
    '6bSU26-2_20200806tp.txt',
    '6bSU26-2_20201118tip.txt',
    '6bSU26-2_20201204tip.txt',
    '6bSU26-2_20210113tip.txt',
    '6bSU26-2_20210216tip.txt',
    '6bSU26-2_20210316tip.txt',
    '6bSU26-2_20210529tp.txt',
    '6bSU26-2_20211105tip.txt',
    '6bSU26-2_20220521tp.txt',
    '6bSU26-2_20221110tip.txt',
    '6bSU26-2_20230221tip.txt',
    '6bSU26-2_20230506tp.txt',
    '6bSU26-2_20231004tip.txt',
    '6bSU26-2_20231127tpsp.txt',
    '6bSU26-3_20111215tip.txt',
    '6bSU26-3_20120806tip.txt',
    '6bSU26-3_20120806tp.txt',
    '6bSU26-3_20121031tip.txt',
    '6bSU26-3_20130523tip.txt',
    '6bSU26-3_20130822tp.txt',
    '6bSU26-3_20131203tip.txt',
    '6bSU26-3_20140304tip.txt',
    '6bSU26-3_20140810tp.txt',
    '6bSU26-3_20141027tip.txt',
    '6bSU26-3_20150307tip.txt',
    '6bSU26-3_20150708tp.txt',
    '6bSU26-3_20160114tip.txt',
    '6bSU26-3_20160722tp.txt',
    '6bSU26-3_20161116tip.txt',
    '6bSU26-3_20170726tp.txt',
    '6bSU26-3_20171107tip.txt',
    '6bSU26-3_20180306tip.txt',
    '6bSU26-3_20180518tp.txt',
    '6bSU26-3_20181010tip.txt',
    '6bSU26-3_20190322tip.txt',
    '6bSU26-3_20190606tp.txt',
    '6bSU26-3_20190929tip.txt',
    '6bSU26-3_20200129tip.txt',
    '6bSU26-3_20200807tp.txt',
    '6bSU26-3_20201118tip.txt',
    '6bSU26-3_20210316tip.txt',
    '6bSU26-3_20210826tp.txt',
    '6bSU26-3_20211105tip.txt',
    '6bSU26-3_20220517tp.txt',
    '6bSU26-3_20221110tip.txt',
    '6bSU26-3_20230221tip.txt',
    '6bSU26-3_20230505tp.txt',
    '6bSU26-3_20231004tip.txt'
]

desired_reg_ids = [
    '6b01385', '6b01384', '6b01383', '6b01382', '6b01354', '6b01350', '6b01346', '6b01342', '6b01338',
    '6b01334', '6b01330', '6b01323', '6b01319', '6b01315', '6b01310', '6b01306', '6b01302', '6b01298',
    '6b01294', '6b01287', '6b01283', '6b01277', '6b01267', '6b01263', '6b01257', '6b01253', '6b01247',
    '6b01243', '6b01237', '6b01233', '6b01227', '6b01220', '6b01186', '6b01182', '6b01179', '6b01175'
]

# Function to extract data from each file
def process_file(file_path):
    # Extract date from file name using regular expression
    match = re.search(r'_(\d{4})(\d{2})(\d{2})(?:tp|tip)\.txt$', file_path)
    if match:
        year = int(match.group(1))
        month = int(match.group(2))
        day = int(match.group(3))
        date_formatted = f'{day:02d}/{month:02d}/{year}'  # Format as DD/MM/YYYY

        # Read data from file and extract relevant information
        data = []
        with open(file_path, 'r') as file:
            lines = file.readlines()
            for line in lines:
                parts = line.strip().split('\t')  # Assuming tab-separated values
                if len(parts) >= 7:  # Ensure there are at least 7 parts (adjust if necessary)
                    reg_id = parts[6].strip()  # Assuming Reg_ID is in the 7th column
                    try:
                        chainage = float(parts[3].strip())  # Assuming Chainage is in the 4th column
                        elevation = float(parts[2].strip())  # Assuming Elevation is in the 3rd column
                        if reg_id in desired_reg_ids:
                            data.append({
                                'Site': 'Start Bay',
                                'Reg_ID': reg_id,
                                'Date': date_formatted,
                                'Chainage': chainage,
                                'Elevation': elevation
                            })
                    except ValueError as e:
                        print(f'Error parsing line: {line.strip()} in file {file_path}: {e}')
                else:
                    print(f'Unexpected line format: {line.strip()} in file {file_path}')
        
        return data
    else:
        print(f'Invalid file name format: {file_path}')
        return []

# Process each file and collect data
all_data = []
for file_name in file_list:
    file_path = os.path.join(data_folder, file_name)
    file_data = process_file(file_path)
    all_data.extend(file_data)

# Create DataFrame
df = pd.DataFrame(all_data)

# Output to CSV with DD/MM/YYYY date format
output_file = os.path.join(data_folder, 'StartBay_Profiles.csv')
df.to_csv(output_file, index=False, date_format='%d/%m/%Y')

print(f'CSV file successfully created at: {output_file}')

```
