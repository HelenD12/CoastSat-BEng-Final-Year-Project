```python
import pandas as pd
import os
import re

# Update file path to folder containing data
data_folder = r'C:\Users\helen\Documents\CCO\Perranporth'

# Define file list and desired Reg IDs
file_list = [
    '7a7A3-13_20070401tp.txt',
'7a7A3-13_20080221tip.txt',
'7a7A3-13_20100227tip.txt',
'7a7A3-13_20110305tip.txt',
'7a7A3-13_20110928tip.txt',
'7a7A3-13_20120604tp.txt',
'7a7A3-13_20120919tip.txt',
'7a7A3-13_20130114tip.txt',
'7a7A3-13_20140303tip.txt',
'7a7A3-13_20150323tip.txt',
'7a7A3-13_20160313tip.txt',
'7a7A3-13_20170211tp.txt',
'7a7A3-13_20170211tstr.txt',
'7a7A3-13_20180111tpsp.txt',
'7a7A3-13_20180306tip.txt',
'7a7A3-13_20190124tip.txt',
'7a7A3-13_20200313tip.txt',
'7a7A3-13_20210214tip.txt',
'7a7A3-13_20220306tp.txt',
'7a7A3-13_20230322tip.txt',
'7a7A3-15_20070406tp.txt',
'7a7A3-15_20071027tp.txt',
'7a7A3-15_20071203tpsp.txt',
'7a7A3-15_20080221tip.txt',
'7a7A3-15_20081015tip.txt',
'7a7A3-15_20090329tip.txt',
'7a7A3-15_20090922tip.txt',
'7a7A3-15_20100317tip.txt',
'7a7A3-15_20100401tpsp.txt',
'7a7A3-15_20100911tip.txt',
'7a7A3-15_20110324tip.txt',
'7a7A3-15_20110928tip.txt',
'7a7A3-15_20120321tp.txt',
'7a7A3-15_20120920tip.txt',
'7a7A3-15_20130114tip.txt',
'7a7A3-15_20130919tip.txt',
'7a7A3-15_20140112tpsp.txt',
'7a7A3-15_20140131tip.txt',
'7a7A3-15_20140912tip.txt',
'7a7A3-15_20150218tip.txt',
'7a7A3-15_20150830tip.txt',
'7a7A3-15_20160313tip.txt',
'7a7A3-15_20161017tip.txt',
'7a7A3-15_20170212tp.txt',
'7a7A3-15_20170212tstr.txt',
'7a7A3-15_20171008tip.txt',
'7a7A3-15_20180205tip.txt',
'7a7A3-15_20181009tip.txt',
'7a7A3-15_20190124tip.txt',
'7a7A3-15_20190902tip.txt',
'7a7A3-15_20200313tip.txt',
'7a7A3-15_20200919tip.txt',
'7a7A3-15_20210213tip.txt',
'7a7A3-15_20210909tip.txt',
'7a7A3-15_20220203tp.txt',
'7a7A3-15_20220928tip.txt',
'7a7A3-15_20230309tip.txt',
'7a7A3-15_20230916tip.txt',
'7a7A3-17_20070404tp.txt',
'7a7A3-17_20070927tp.txt',
'7a7A3-17_20080208tip.txt',
'7a7A3-17_20080929tip.txt',
'7a7A3-17_20090210tip.txt',
'7a7A3-17_20090918tip.txt',
'7a7A3-17_20100302tip.txt',
'7a7A3-17_20100908tip.txt',
'7a7A3-17_20110218tip.txt',
'7a7A3-17_20110927tip.txt',
'7a7A3-17_20120920tp.txt',
'7a7A3-17_20121215tip.txt',
'7a7A3-17_20130326tip.txt',
'7a7A3-17_20131205tip.txt',
'7a7A3-17_20140112tpsp.txt',
'7a7A3-17_20140303tip.txt',
'7a7A3-17_20140910tip.txt',
'7a7A3-17_20150219tip.txt',
'7a7A3-17_20151001tip.txt',
'7a7A3-17_20160209tip.txt',
'7a7A3-17_20161018tip.txt',
'7a7A3-17_20170228tp.txt',
'7a7A3-17_20171008tip.txt',
'7a7A3-17_20180205tip.txt',
'7a7A3-17_20181009tip.txt',
'7a7A3-17_20190124tip.txt',
'7a7A3-17_20190902tip.txt',
'7a7A3-17_20200312tip.txt',
'7a7A3-17_20200919tip.txt',
'7a7A3-17_20210213tip.txt',
'7a7A3-17_20210909tip.txt',
'7a7A3-17_20220202tp.txt',
'7a7A3-17_20220928tip.txt',
'7a7A3-17_20230222tip.txt',
'7a7A3-17_20230916tip.txt',
'7a7A3-18_20070504tp.txt',
'7a7A3-18_20070926tp.txt',
'7a7A3-18_20080207tip.txt',
'7a7A3-18_20080930tip.txt',
'7a7A3-18_20090212tip.txt',
'7a7A3-18_20090921tip.txt',
'7a7A3-18_20100415tip.txt',
'7a7A3-18_20101107tip.txt',
'7a7A3-18_20110322tip.txt',
'7a7A3-18_20110928tip.txt',
'7a7A3-18_20120323tp.txt',
'7a7A3-18_20121115tip.txt',
'7a7A3-18_20130327tip.txt',
'7a7A3-18_20140302tip.txt',
'7a7A3-18_20150220tip.txt',
'7a7A3-18_20160506tip.txt',
'7a7A3-18_20170428tp.txt',
'7a7A3-18_20180218tip.txt',
'7a7A3-18_20190125tip.txt',
'7a7A3-18_20200312tip.txt',
'7a7A3-18_20210212tip.txt',
'7a7A3-18_20220303tp.txt',
'7a7A3-18_20230222tip.txt',
'7a7A3-2_20070930tp.txt',
'7a7A3-2_20080210tip.txt',
'7a7A3-2_20081016tp.txt',
'7a7A3-2_20090301tip.txt',
'7a7A3-2_20090908tip.txt',
'7a7A3-2_20100203tip.txt',
'7a7A3-2_20101108tip.txt',
'7a7A3-2_20110320tip.txt',
'7a7A3-2_20110928tip.txt',
'7a7A3-2_20120311tp.txt',
'7a7A3-2_20120918tip.txt',
'7a7A3-2_20130228tip.txt',
'7a7A3-2_20130920tip.txt',
'7a7A3-2_20140202tip.txt',
'7a7A3-2_20140910tip.txt',
'7a7A3-2_20150220tip.txt',
'7a7A3-2_20150831tip.txt',
'7a7A3-2_20160213tpsp.txt',
'7a7A3-2_20160312tip.txt',
'7a7A3-2_20160312tip.txt',
'7a7A3-2_20161019tip.txt',
'7a7A3-2_20170214tp.txt',
'7a7A3-2_20171010tip.txt',
'7a7A3-2_20180203tip.txt',
'7a7A3-2_20181008tip.txt',
'7a7A3-2_20190125tip.txt',
'7a7A3-2_20190904tip.txt',
'7a7A3-2_20200112tip.txt',
'7a7A3-2_20200918tip.txt',
'7a7A3-2_20210115tip.txt',
'7a7A3-2_20210909tip.txt',
'7a7A3-2_20220301tp.txt',
'7a7A3-2_20220926tip.txt',
'7a7A3-2_20230307tip.txt',
'7a7A3-2_20230915tip.txt',
'7a7A3-4_20070930tp.txt',
'7a7A3-4_20080210tip.txt',
'7a7A3-4_20081016tip.txt',
'7a7A3-4_20090313tip.txt',
'7a7A3-4_20090908tip.txt',
'7a7A3-4_20100203tip.txt',
'7a7A3-4_20101108tip.txt',
'7a7A3-4_20110320tip.txt',
'7a7A3-4_20110928tip.txt',
'7a7A3-4_20120310tp.txt',
'7a7A3-4_20120918tip.txt',
'7a7A3-4_20130228tip.txt',
'7a7A3-4_20130920tip.txt',
'7a7A3-4_20140202tip.txt',
'7a7A3-4_20140910tip.txt',
'7a7A3-4_20150220tip.txt',
'7a7A3-4_20150831tip.txt',
'7a7A3-4_20160312tip.txt',
'7a7A3-4_20161019tip.txt',
'7a7A3-4_20170210tp.txt',
'7a7A3-4_20171010tip.txt',
'7a7A3-4_20180203tip.txt',
'7a7A3-4_20181012tip.txt',
'7a7A3-4_20190124tip.txt',
'7a7A3-4_20190902tip.txt',
'7a7A3-4_20200112tip.txt',
'7a7A3-4_20200918tip.txt',
'7a7A3-4_20210115tip.txt',
'7a7A3-4_20210909tip.txt',
'7a7A3-4_20220321tp.txt',
'7a7A3-4_20220926tip.txt',
'7a7A3-4_20230307tip.txt',
'7a7A3-4_20230915tip.txt',
'7a7A3-8_20070404tp.txt',
'7a7A3-8_20071012tp.txt',
'7a7A3-8_20071203tpsp.txt',
'7a7A3-8_20080211tip.txt',
'7a7A3-8_20080903tip.txt',
'7a7A3-8_20090309tip.txt',
'7a7A3-8_20090919tip.txt',
'7a7A3-8_20100318tip.txt',
'7a7A3-8_20100401tpsp.txt',
'7a7A3-8_20101009tip.txt',
'7a7A3-8_20110220tip.txt',
'7a7A3-8_20111124tip.txt',
'7a7A3-8_20120608tp.txt',
'7a7A3-8_20121015tip.txt',
'7a7A3-8_20130329tip.txt',
'7a7A3-8_20130921tip.txt',
'7a7A3-8_20140113tpsp.txt',
'7a7A3-8_20140301tip.txt',
'7a7A3-8_20140912tip.txt',
'7a7A3-8_20150218tip.txt',
'7a7A3-8_20150831tip.txt',
'7a7A3-8_20160308tip.txt',
'7a7A3-8_20161018tip.txt',
'7a7A3-8_20170301tp.txt',
'7a7A3-8_20171009tip.txt',
'7a7A3-8_20171024tpsp.txt',
'7a7A3-8_20180105tpsp.txt',
'7a7A3-8_20180306tip.txt',
'7a7A3-8_20181008tip.txt',
'7a7A3-8_20190123tip.txt',
'7a7A3-8_20190904tip.txt',
'7a7A3-8_20200305tpsp.txt',
'7a7A3-8_20200311tip.txt',
'7a7A3-8_20200919tip.txt',
'7a7A3-8_20210115tip.txt',
'7a7A3-8_20210908tip.txt',
'7a7A3-8_20220304tp.txt',
'7a7A3-8_20220928tip.txt',
'7a7A3-8_20230224tip.txt',
'7a7A3-8_20230916tip.txt',
'7a7A3-9_20070406tp.txt',
'7a7A3-9_20071203tpsp.txt',
'7a7A3-9_20080211tip.txt',
'7a7A3-9_20090310tip.txt',
'7a7A3-9_20100216tip.txt',
'7a7A3-9_20110406tip.txt',
'7a7A3-9_20111124tip.txt',
'7a7A3-9_20120606tp.txt',
'7a7A3-9_20121015tip.txt',
'7a7A3-9_20130330tip.txt',
'7a7A3-9_20140113tpsp.txt',
'7a7A3-9_20140301tip.txt',
'7a7A3-9_20150218tip.txt',
'7a7A3-9_20160308tip.txt',
'7a7A3-9_20170301tp.txt',
'7a7A3-9_20180306tip.txt',
'7a7A3-9_20190123tip.txt',
'7a7A3-9_20200311tip.txt',
'7a7A3-9_20210115tip.txt',
'7a7A3-9_20220305tp.txt',
'7a7A3-9_20230224tip.txt', 
]

desired_reg_ids = [
    '7a01435', '7a01438', '7a01441', '7a01444', '7a01448', '7a01454', '7a01464', '7a01477', '7a01487', '7a01497'
]


def process_file(file_path):
    # Extract date from file name using regular expression
    match = re.search(r'_(\d{4})(\d{2})(\d{2})(?:tp|tip)\.txt$', file_path)
    if match:
        year = int(match.group(1))
        month = int(match.group(2))
        day = int(match.group(3))
        date_formatted = f'{day:02d}/{month:02d}/{year}'  # Format as DD/MM/YYYY

        # Read data from file and extract relevant information
        data = []
        try:
            with open(file_path, 'r') as file:
                lines = file.readlines()
                for line in lines:
                    parts = line.strip().split('\t')  # Assuming tab-separated values
                    if len(parts) >= 7:  # Ensure there are at least 7 parts (adjust if necessary)
                        try:
                            reg_id = parts[6].strip()  # Assuming Reg_ID is in the 7th column
                            chainage = float(parts[3].strip())  # Assuming Chainage is in the 4th column
                            elevation = float(parts[2].strip())  # Assuming Elevation is in the 3rd column
                            if reg_id in desired_reg_ids:
                                data.append({
                                    'Site': 'Perranporth',
                                    'Reg_ID': reg_id,
                                    'Date': date_formatted,
                                    'Chainage': chainage,
                                    'Elevation': elevation
                                })
                        except (ValueError, IndexError) as e:
                            print(f'Error parsing line: {line.strip()} in file {file_path}: {e}')
                    else:
                        print(f'Unexpected line format: {line.strip()} in file {file_path}')
        except FileNotFoundError:
            print(f'File not found: {file_path}')
        
        return data
    else:
        print(f'Invalid file name format: {file_path}')
        return []

# Process each file and collect data
all_data = []
for file_name in file_list:
    file_path = os.path.join(data_folder, file_name)
    file_data = process_file(file_path)
    all_data.extend(file_data)

# Create DataFrame
df = pd.DataFrame(all_data)

# Output to CSV with DD/MM/YYYY date format
output_file = os.path.join(data_folder, 'Perranporth_Profiles.csv')
df.to_csv(output_file, index=False, date_format='%d/%m/%Y')

print(f'CSV file successfully created at: {output_file}')
```
